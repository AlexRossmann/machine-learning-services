# data analysis
let's start by looking at the given data sets

```python
# at first let's have a look at the "train data" or in this notebook also called result data
data_train
```

```python
# first set no to every NaN in passed_visual_inspection due to not appeared in inspection after cancelling the milling
data_train["passed_visual_inspection"] = data_train["passed_visual_inspection"].fillna("no")

# then we use the map function to set 0 and 1 to the ordinal features, in new columns
data_train["dummie_tool_condition"] = data_train["tool_condition"].map({"worn":"1", "unworn":"0"})
data_train["dummie_machining_finalized"] = data_train["machining_finalized"].map({"yes":"1", "no":"0"})
data_train["dummie_passed_visual_inspection"] = data_train["passed_visual_inspection"].map({"yes":"1", "no":"0"})
data_train
```

```python
# lets check the experiment data files for missing values
exp_missings = []
for i in range(18):
    i += 1

    #load the experiment csv data from the directory, no preceeded zero needed if experiment number is bigger than 9
    if(i < 10):
        file_name = "Toolwear/experiment_0" + str(i)
    else:
        file_name = "Toolwear/experiment_" + str(i)
    file_name = file_name + ".csv"

    experiment = pd.read_csv(file_name)
    print(str(i) + ": " + str(experiment.shape[0]) )
    # make a tuple with the file name and the columns with missing values
    exp_missings.append( (file_name, experiment.columns[experiment.isnull().any()].tolist() )  )
    
print(exp_missings)
```

to use the datasets we match them together with the result row of the experiment
for this we use the number of the exmperiment

```python
data_frames = []
for i in range(18):
    i += 1
    
    #get the experiments result row, conditioned by the experiment Number
    data_train_row = data_train[data_train["No"] == i]
    
    #load the experiment csv data from the directory, no preceeded zero needed if experiment number is bigger than 9
    if(i < 10):
        file_name = "Toolwear/experiment_0" + str(i)
    else:
        file_name = "Toolwear/experiment_" + str(i) 
    file_name = file_name + ".csv"
    
    experiment = pd.read_csv(file_name)
    print(experiment.count("columns"))
    experiment["exp_No"] = data_train_row.iloc[0]["No"]
    experiment["feedrate"] = data_train_row.iloc[0]["feedrate"]
    experiment["clamp_pressure"] = data_train_row.iloc[0]["clamp_pressure"]
    experiment["dummie_tool_condition"] = data_train_row.iloc[0]["dummie_tool_condition"]
    experiment["dummie_passed_visual_inspection"] = data_train_row.iloc[0]["dummie_passed_visual_inspection"]
    
    data_frames.append(experiment)
```

```python
# create a new dataframe by concatting all the frames in the list
data_concat = pd.concat(data_frames)
```

# Lets train the models

```python
# get the value for the X alignment to predict from 
x_values = data_concat.drop(["dummie_passed_visual_inspection"], axis=1)
# get the values for the Y alignment which are our targets
y_values = data_concat["dummie_passed_visual_inspection"]
print(x_values.shape)
print(y_values.shape)

# now lets split them to training 75 and testing 25 data
x_training, x_testing, y_training, y_testing = train_test_split(x_values, y_values, test_size=0.25) 
print(y_testing)
print(y_training)
```

## Random forest

```python
rndF = RandomForestClassifier(n_estimators=100, random_state=42)
rndF.fit(x_training, y_training)
```

```python
# Vorhersage auf den Testdaten
rndF_predict = rndF.predict(x_testing)

# Evaluierung der Modellleistung
accuracy = accuracy_score(y_testing, rndF_predict)
accuracy
```

# nearest neighbour

```python
def evaluate_knn(X_train, X_valid, y_train, y_valid, k):
    
    nb_model = KNeighborsClassifier(n_neighbors=k)
    nb_model.fit(X_train, y_train)

    nb_model_pred = nb_model.predict(X_valid)
    accuracy = accuracy_score(y_valid, nb_model_pred)
    return accuracy

# take a range of k as neighbours
k_values = list(range(1, 25))
accuracies = []

# now iterate through all ks and get the accuracy
for k in k_values:
    accuracy = evaluate_knn(x_training, x_testing, y_training, y_testing, k)
    accuracies.append(accuracy)
    
# lets have a look at the results
plt.plot(k_values, accuracies, marker='o')
plt.title('Ellenbogen-Methode für k-NN')
plt.xlabel('Anzahl der Nachbarn (k)')
plt.ylabel('Genauigkeit')
plt.show()

optimal_k_index = accuracies.index(max(accuracies))
print(optimal_k_index)
```

```python
nb_model = KNeighborsClassifier(6)
nb_model.fit(x_training, y_training)

nb_model_pred = nb_model.predict(x_testing)
accuracy = accuracy_score(y_testing, nb_model_pred)
print(accuracy)
```
